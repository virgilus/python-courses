## Introduction

Pytest is a powerful and flexible testing framework for Python. It allows you to create, organize and execute tests in a simple and efficient way. In this course, we will explore the basics of Pytest, learn to write tests, and discover some advanced features.

## Installing Pytest

To use Pytest, you must first install it.

```sh
pip install pytest
```

## Writing Simple Tests

### Basic Test Structure

A Pytest test is simply a Python function that starts with `test_`. Go to the "project" folder. You will find two folders there:

- "my_lib" which is a library containing a module named "my_functions".
- "tests" which will contain the tests that we will give to pytest.

## Using Assertions

Assertions are used to verify that test results are correct. Pytest supports several types of assertions.

### Basic Assertions

````python
def test_assertions():
    assert 1 == 1
    assert 2 + 2 == 4
    assert "hello".upper() == "HELLO"
````

### Assertions with Messages

You can add messages to assertions to provide additional information in case of failure.

````python
def test_assertions_with_messages():
    assert 1 == 1, "1 should be equal to 1"
    assert 2 + 2 == 4, "2 + 2 should be equal to 4"
    assert "hello".upper() == "HELLO", "hello.upper() should be equal to HELLO"
````

## Running Pytest

When you type the following command in a terminal:
```sh
pytest
```

Pytest examines all files in the current folder as well as all folders below it. If it finds files with the ".py" extension whose name starts with "test_", it will examine and run them.
