# ‚öôÔ∏è FUNCTIONS

üéØ We have already used many different functions so far. However, in programming it is common to create your own functions that will be perfectly adapted to your needs. Python allows you to do this very easily in just a few lines!

## Syntax

üëâ To do this we use the following syntax, which is very similar to that of loops or tests:

- First, write the keyword `def` to make Python understand that the following will be a function.
- Then type name you want to give to your function.
- Add parentheses between which you can indicate arguments.
- Add the sign `:` to specify that the sequence will be the content of the function.
- At any time during the function we can tell it that we want it to return a value using the `return` statement. But this is not mandatory.

````python
def my_first_function():
    return "First function!"
````

## Execution

ü§® Nothing happens when you run the cell? This is normal, your function is now **defined**, it exists somewhere in Python's memory. To execute it, you just have to call it like any other function.

````python
my_first_function()
````

## Parameters

üìñ When defining a function you can define parameters by indicating the names of variables inside the parenthesis. The following function, which returns the square of a number, has one parameter: "x".

```python
def square(x):
    return x * x
```

## Arguments

Let's run the function with the parameter "x" set to 2:

```python
square(2)
```
üìñ Here **the number 2 is the argument** given to the function (and "x" is the parameter). The argument will pass the value to the function's parameter.
Let's test this function by giving it different arguments.

````python
def square(x):
    return x * x

print(square(2))
print(square(7))
print(square(9))
````

üëâ When we **call** (execute / run) the function, we pass the desired value as input. This value then takes the place of "x" in our function. But notice that we could have written this and obtained exactly the same results:

````python
def square(blablabla):
    return blablabla * blablabla

print(square(2))
print(square(7))
print(square(9))

# We can also use a variable as argument
n = 49
print(square(n))
````

## The `return` statement

üìñ The vast majority of functions have one or more `return` statements. This statement manages the output parameters, so it indicates what the function returns, what our function "transforms" into. After executing a `return` the function automatically terminates without executing the other lines of code.

In the following example the function returns the square of a number if given a negative integer. If the number is between 0 and 100 inclusive it returns the value of the number plus 10.

Otherwise it returns nothing which is a special object in Python called `None`.

````python
def square_or_plus_ten(x):
    if x < 0: return x ** 2
    elif x <= 100: return x + 10
````

````python
square_or_plus_ten(-4)
````

````python
square_or_plus_ten(50)
````

````python
square_or_plus_ten(400)
````

## Multiple parameters

üìñ You can give different parameters to a function by separating them with a comma. For example the following function returns different numbers multiplied by each other:

````python
def mult(a, b, c):
    return a * b * c

mult(3, 4, 5)
````

## Default Arguments

üìñ If the user does not specify any arguments, the default value of the parameter can be set using the `=` sign.

‚ùóÔ∏èDefault parameters must be declared **after** non-default parameters (mandatory parameters).

**Note**: By convention, whitespace are not used on either side of the equal operator (`=`) in the definition of a function or when calling it (PEP8).

Ex:

````python
def multiply_by_10(x=10):
    return x * 10

print(multiply_by_10())
print(multiply_by_10(5))
print(multiply_by_10(23))
print(multiply_by_10())
````

## Keyword Arguments vs Positional Arguments

When **calling a function**, you can provide arguments to the different parameters:

- üëâ Either by calling them in the order they were declared in the function: **Positional Argument**.
- üëâ Or by calling them by their name: **Keyword Argument**.
- üëâ Or by mixing these two methods. In this case, you must always **start with the positional arguments** and then call the keyword arguments.

````python
def multiply(a, b, text):
    return text + str(a * b)

# 1 - Only positional arguments
print(multiply(6, 2, "Results = "))

# 2 - Only keywords arguments
print(multiply(text="Results = ", b=6, a=2))

# 3 - Both methods
print(multiply(2, text="Results = ", b=6))
````

## Type Hint

In Python, a type hint is a feature that allows you to specify the expected function arguments, and return values. Type hints are not enforced at runtime, but they serve as documentation and can be used by verification tools and IDEs to catch potential errors before the code is executed.

To add a type hint, use ":" after your parameter. You can use the pipe ```|``` to add several types.

**Note**: To access advanced features of *type hints*, you can use the `typing` library.

````python
def add(a: int, b: int | float) -> str:
    return f"The result is : {a + b}"
````

## ```*args``` et ```**kwargs```

In Python, ```*args``` and ```**kwargs``` are used to pass a variable number of arguments to a function. They allow you to define functions that can accept an arbitrary number of positional and keyword arguments, respectively. Both names are just conventions, they are not a python reserved keywords.


### *args

It is used to pass a variable number positional arguments to a function. The arguments are passed as a tuple.

````python
def demo_args(*args):
    return args

demo_args(2, 3, 4, 12)
````

### ```**kwargs```
```**kwargs``` is used to pass a variable number of keyword arguments to a function. The arguments are passed as a dictionary.

````python
def demo_kwargs(**kwargs):
    return kwargs

demo_kwargs(a=2, b=3, c=4, d=12)
````

````python
# And you mix both of course!

def demo_args_and_kwargs(*args, **kwargs):
    return args, kwargs

demo_args_and_kwargs(100, 101, 102, "banana", a=2, b=3, c=4, d=12)
````

## Signature

In Python, a function signature refers to the definition of a function, including its name, parameters, type hint and any default values for those parameters. The function signature provides a clear and concise way to understand what inputs a function expects and how it should be called. Ex:

```python
add(a: int, b: int | float) -> str
```
