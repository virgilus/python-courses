## Introduction

It can sometimes be useful to create graphical user interfaces (GUIs) in Python. Graphical interfaces allow users to interact with programs in an intuitive and visual way. The two most popular graphical APIs in this area are Tkinter and PyQt6. We will only focus on Tkinter here.

## Tkinter

Tkinter is the standard library for creating graphical interfaces in Python. It is included with the standard Python distribution, which makes it easily accessible. Tkinter is simple to use and ideal for small to medium-sized projects.

[source](https://likegeeks.com/python-gui-examples-tkinter-tutorial/#Create_your_first_GUI_application)

```python
from tkinter import *

window = Tk()
window.title("Welcome to our Python app!")

window.mainloop()
```

```python
from tkinter import *

window = Tk()

window.title("Welcome to our Python app!")

lbl = Label(window, text="Hello") # Adding a label that says Hello

lbl.grid(column=0, row=0)

window.mainloop()
```

```python
from tkinter import *

window = Tk()
window.title("Welcome to our Python app!")

window.geometry('700x400') # setting the window size
lbl = Label(window, text="Hello")
lbl.grid(column=0, row=0) # using the grid as a referential
btn = Button(window, text="Click Me")
btn.grid(column=1, row=0)

window.mainloop()
```

```python
from tkinter import *

window = Tk()

window.title("Welcome to our Python app!")
window.geometry('700x400')
lbl = Label(window, text="Hello")
lbl.grid(column=0, row=0)

def clicked(): # we can configure function inside our tkinter app
    lbl.configure(text="Button was clicked !!")

btn = Button(window, text="Click Me", command=clicked)
btn.grid(column=1, row=0)

window.mainloop()
```
